```python
from ultralytics import YOLO
import torch

device = 0 if torch.cuda.is_available() else "cpu"

# training
model = YOLO("yolo11s-seg.pt")

model.train(
    data="/kaggle/working/data.yaml",
    epochs=100,
    imgsz=640,
    batch=8,
    patience=40,
    device=device
)
```

```python
# validation
from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/segment/train/weights/best.pt")

metrics = model.val(
    data="/kaggle/working/data.yaml",
    split="val"
)

print(metrics)
```

```python
# testing
model.predict(
    source="/kaggle/input/datasets/jasminesingh01/yolo11segdata/test/images",
    save=True,
    conf=0.25
)
```
